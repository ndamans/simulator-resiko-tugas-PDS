Langkah 1: Pelatihan Model dan Scaler (Fase Pengembangan)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import joblib

# 1. Menyiapkan data historis (Baseline)
# Fitur: [Suhu_Mesin, Getaran_Mesin]
X_train = np.array([[60, 2], [70, 4], [80, 6], [90, 8], [100, 10]])
y_train = np.array([10, 25, 45, 70, 95]) # Skor Risiko

# 2. Preprocessing (Scaler)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# 3. Training Model
model = LinearRegression().fit(X_train_scaled, y_train)
print("Model dan Scaler berhasil dilatih pada data historis.")

# Menyimpan model dan scaler ke dalam file .pkl atau .joblib
joblib.dump(model, 'model_risiko_v1.joblib')
joblib.dump(scaler, 'scaler_risiko_v1.joblib')
print("File 'model_risiko_v1.joblib' dan 'scaler_risiko_v1.joblib' telah tersimpan.")
```

Langkah 3: Operasionalisasi (Model Loading & Inference)

```python
# 1. Memuat kembali objek dari disk
loaded_model = joblib.load('model_risiko_v1.joblib')
loaded_scaler = joblib.load('scaler_risiko_v1.joblib')

# 2. Data baru dari sensor (Real-time)
# [Suhu, Getaran]
data_baru = np.array([[85, 7]])

# 3. Transformasi dan Prediksi (Inference)
data_baru_scaled = loaded_scaler.transform(data_baru)
hasil_simulasi = loaded_model.predict(data_baru_scaled)
```

Langkah 4: Implementasi Deteksi Drift Sederhana

```python
def check_data_drift(new_data, train_mean, threshold=2.0):
  # Menghitung selisih rata-rata input baru terhadap rata-rata latihan
  drift = np.abs(np.mean(new_data) - np.mean(train_mean))
  if drift > threshold:
    return f"WARNING: Terdeteksi Drift sebesar {drift:.2f}. Segera kalibrasi model!"
  else:
    return "Sistem Stabil: Data input sesuai dengan profil model."

# Rata-rata data latihan sebagai patokan
train_mean_ref = np.mean(X_train)
# Contoh data sensor yang mengalami gangguan (anomali)
data_anomali = np.array([[200, 50]])

print(check_data_drift(data_anomali, train_mean_ref))
```

Langkah 5: Praktik Etika (Anonymization)

```python
def clean_sensitive_data(df_input):
  # Menghapus kolom sensitif (PII - Personally Identifiable Information)
  cols_to_remove = ['Nama_Operator', 'NIK_Petugas']
  df_cleaned = df_input.drop(columns=[c for c in cols_to_remove if c in df_input.columns])
  return df_cleaned

# Contoh input data kotor
raw_input = pd.DataFrame({
  'Nama_Operator': ['Budi'],
  'Suhu': [85],
  'Getaran': [7]
})

final_input = clean_sensitive_data(raw_input)
print("\nInput setelah Anonymization (Etika Data):")
print(final_input)
```
